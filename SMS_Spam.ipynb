```python
from tensorflow.keras.layers import TextVectorization

text_vecorizer = TextVectorization(max_tokens=None,
                                   standardize='lower_and_strip_punctuation',
                                   split='whitespace',
                                   ngrams=None,
                                   output_mode="int",
                                   output_sequence_length=None)
```

```python
from sklearn.model_selection import train_test_split
train_sentence, val_sentence, train_label, val_label = train_test_split(train_df_shuffled["text"].to_numpy(),
                                                                            train_df_shuffled["label"].to_numpy(),
                                                                            test_size=0.1, # dedicate 10% of samples to validation set
                                                                            random_state=42)
```

```python
max_vocab_length = 100000
max_length = 100
text_vectorizer = TextVectorization(max_tokens=max_vocab_length,
    standardize="lower_and_strip_punctuation",
    split="whitespace",
    ngrams=None,
    output_mode="int",
    output_sequence_length=max_length,
    pad_to_max_tokens=False,
    vocabulary=None,
    idf_weights=None,
    sparse=False,
    ragged=False,
    encoding="utf-8",
    name=None)
```

```python
text_vectorizer.adapt(train_sentence)
```

```python
sample_text = 'رول چایی خوب بود گاتا طعم و مزه جالبی نداشت و به نظر تازه هم نبود'
text_vectorizer([sample_text])
```

```python
word_in_vocab = text_vectorizer.get_vocabulary()
top_5 = word_in_vocab[:5]
last_5 = word_in_vocab[-5:]
print(top_5)
print(last_5)
```

```python
embedding = tf.keras.layers.Embedding(input_dim=max_vocab_length,
                                      output_dim=128,
                                      embeddings_initializer="uniform",
                                      input_length=max_length)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

model0 = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])

model0.fit(train_sentence, train_label)
```

```python
baseline_score = model0.score(val_sentence, val_label)
baseline_score
```

```python
baseline_predict = model0.predict(val_sentence)
print(f'The baseline predict is: {baseline_predict[:10]}\n')
print(f'The real label is: {val_label[:10]}')
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def calculate_result(y_true, y_pred):
  model_accuracy = accuracy_score(y_true, y_pred)
  model_precision, model_recall, f1_model, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
  model_result = {'Accuracy':model_accuracy,
                  'Precision':model_precision,
                  'Recall':model_recall,
                  'F1_score':f1_model}
  return model_result
```

```python
calculate_result(y_true=val_label,
                 y_pred=baseline_predict)
```

## Model 1: A simple dense model

```python
tf.random.set_seed(42)

inputs = tf.keras.layers.Input(shape=(1, ), dtype='string')
x = text_vectorizer(inputs)
x = embedding(x)
x = tf.keras.layers.GlobalAveragePooling1D()(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model1 = tf.keras.Model(inputs, outputs)

model1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=['accuracy'])
model1.fit(train_sentence,
          train_label,
          epochs=5,
          validation_data=(val_sentence, val_label))
```

```python
model1_pred_prob = model1.predict(val_sentence)
model1_pred_prob[:10]
```

```python
model1_predict = tf.squeeze(tf.round(model1_pred_prob))
model1_predict
```

```python
model_1_results = calculate_result(y_true=val_label,
                                    y_pred=model1_predict)
model_1_results
```

```python
inputs = tf.keras.layers.Input(shape=(1, ), dtype='string')
x = text_vectorizer(inputs)
x = embedding(x)
x = tf.keras.layers.LSTM(64)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model2 = tf.keras.Model(inputs, outputs)

model2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=['accuracy'])
model2.fit(train_sentence,
           train_label,
           epochs=5,
           validation_data=(val_sentence, val_label))
```

```python
import tensorflow_hub as hub

embedd = hub.load('https://www.kaggle.com/models/google/nnlm/frameworks/TensorFlow2/variations/de-dim128/versions/1')
```

```python
sample = embedd(['کیفیت صداش خوبه   اینکه چه مدت سالم بمونه رو خدا می دونه'])
sample
```

```python
sentence_encoder_layer = hub.KerasLayer('https://tfhub.dev/google/universal-sentence-encoder/4',
                                        input_shape=[],
                                        dtype=tf.string,
                                        trainable=False,
                                        name='encoder')
```

```python
tf.random.set_seed(42)

model3 = tf.keras.Sequential([
    sentence_encoder_layer,
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
], name='nnlm_model')

model3.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
               metrics=['accuracy'])

model3.fit(train_sentence,
           train_label,
           epochs=10,
           validation_data=(val_sentence, val_label))
```

```python
tf.random.set_seed(42)

inputs = tf.keras.layers.Input(shape=(1, ), dtype='string')
x = text_vectorizer(inputs)
x = embedding(x)
x = tf.keras.layers.GlobalAveragePooling1D()(x)
x = tf.keras.layers.Dense(512, activation='relu')(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model1 = tf.keras.Model(inputs, outputs)

model1.compile(loss=tf.keras.losses.binary_crossentropy,
               optimizer=tf.keras.optimizers.Adam(),
               metrics=['accuracy'])
model1.fit(train_sentence,
          train_label,
          epochs=5,)
```

```python
model_1_pred_prob = model1.predict(val_sentence)
model_1_pred_prob[:10]
```

```python
model_1_predict = tf.squeeze(tf.round(model_1_pred_prob))
model_1_predict
```

```python
import gradio as gr
import tensorflow as tf
from tensorflow.keras.models import load_model
def sentiment_analysis(text):

    model = load_model('my_model.keras')

    predict = model.predict([text])
    return "positive" if predict > 0.5  else "negative"
```
